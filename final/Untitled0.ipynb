```python
def get_removed_claim(claim, stopwords_list):
  claim_split = claim.split(' ')
  claim_list = []
  for c in claim_split:
    if c not in stopwords_list:
      claim_list.append(c)

  return claim_list

def file_word_num(file_name, claim_list):
  file_name = 'articles/'+file_name
  f = open(file_name)
  d = json.load(f)
  s = 0
  start = 0
  end = 0
  for j in range(len(d)):
    for c in claim_list:
      if c in d[j].split(' '):
        s += 1
        if start == 0:
          start = j
          end = j + 1
        else:
          end = j + 1

        # related_context.append(d[j].lower())
  return s, start, end

def get_file(i, data, stopwords_list, data_type):
  # 如果有 claim 的作者名稱的話，就串接字串
  if data[i]['metadata']['claimant'] != None:
    claim = data[i]['metadata']['claimant'] + ' claimed that ' + data[i]['metadata']['claim']
  else:
    claim = data[i]['metadata']['claim']

  claim = claim.lower()
  # 這邊過濾掉一些沒有用的 stop words，比如說是 the, is 之類的無助於判斷的詞
  # 目的是提高訓練以及判斷效率
  # stopwords_list 應該是下載下來的
  claim_list = get_removed_claim(claim, stopwords_list)
  claim_list += ['fake', 'true']
  all_article = []
  max_rank = -1
  select_file = ''

  a_start = 0
  a_end = 0
  for url in data[i]['metadata']['premise_articles']:
    file_name = data[i]['metadata']['premise_articles'][url]
    s, start, end = file_word_num(file_name, claim_list)
    if s > max_rank:
      select_file = file_name
      a_start = start
      a_end = end
      max_rank = s

  file_name = 'articles/'+select_file
  f = open(file_name)
  d = json.load(f)
  all_context = ''
  if a_end == 0:
    for url in data[i]['metadata']['premise_articles']:
      file_name = data[i]['metadata']['premise_articles'][url]
      f = open(file_name)
      d = json.load(f)
      if d != []:
        break
      
  a_end = len(d)
  for j in range(a_start, a_end):
    all_context += ' ' + d[j]
  if data_type == 'test':
    return all_context
  return all_context, data[i]['label']['rating'], data[i]['label']['original_rating']
```

```python
def get_data(file_name, stopwords_list, data_type):
  f = open(file_name)
  data = json.load(f)
  context_df = []
  rating_df = []
  label_df = []
  for i in range(len(data)):
    print(i)
    if data_type == 'test':
      context = get_file(i, data, stopwords_list, data_type)
      context_df.append(context)
      continue
    context, rating, label = get_file(i, data, stopwords_list, data_type)
    context_df.append(context)
    rating_df.append(rating)
    label_df.append(label)
  if data_type == 'test':
    d = {'context': context_df}
    df = pd.DataFrame(d)
    return df
  d = {'context': context_df, 'label': label_df, 'rating': rating_df}
  df = pd.DataFrame(d)
  return df
```

```python
nltk.download('stopwords')
nltk.download('punkt')
stopwords_list = nltk.corpus.stopwords.words('english')
```

```python
def remove_html(text):
  return re.sub(pattern=r'<\w+/?>', repl='',string=text)
def remove_mentioning(text):
  return re.sub(pattern=r'@[^\s]*', repl='',string=text)
def remove_url(text):
  return re.sub(pattern=r'http[^\s]*', repl='', string=text)
def remove_punctuation(text):
  return re.sub(pattern=r"[{0}]".format(string.punctuation),repl='',string=text)
def remove_numbers(text):
  return re.sub(pattern=r'\d+', repl='', string=text)
def remove_stop_words(text):
  return ' '.join([word for word in word_tokenize(text) if word not in stopwords_list])

def remove_non_en_num(text):
  split_text = text.split(' ')
  re_text = ''
  for t in split_text:
    if t.isascii():
      re_text += t + ' '
  return re_text
def process_data(X):
    return (X
          .apply(remove_html)
          .apply(remove_url)
          .apply(remove_mentioning)
          .apply(remove_punctuation)
          .apply(remove_numbers)
          .apply(remove_non_en_num)
          .apply(remove_stop_words)
          )
```

```python
def get_rating(df):
  ra = []
  for i in range(len(df)):
    t = [0, 0, 1]
    if df['rating'][i] == 0:
      t = [1, 0, 0]
    elif df['rating'][i] == 1:
      t = [0, 1, 0]
    ra.append(t)
  df['rating'] = ra
  return df
```

```python
train_text = train_df['context']
train_label = train_df['rating']

test_text = test_df['context']

validation_text = validation_df['context']
validation_label = validation_df['rating']
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import tensorflow as tf

tokenizer = Tokenizer(num_words=10000, oov_token='<OOV>')
tokenizer.fit_on_texts(train_text)
word_index = tokenizer.word_index
index_word = tokenizer.index_word
train_sequences = tokenizer.texts_to_sequences(train_text)
max_len =  max([len(seq) for seq in train_sequences])
max_len = 1500
np.mean([len(seq) for seq in train_sequences])
```

```python
train_sequences = tokenizer.texts_to_sequences(train_text)
val_sequences = tokenizer.texts_to_sequences(validation_text)
test_sequences = tokenizer.texts_to_sequences(test_text)
```

```python
padded_train_sequences = pad_sequences(train_sequences , maxlen = max_len )
padded_val_sequences = pad_sequences(val_sequences , maxlen = max_len )
padded_test_sequences = pad_sequences(test_sequences , maxlen = max_len ) 
```

```python
a = []
for i in validation_label:
  a.append(i)
validation_label = np.asarray(a)
a = []
for i in train_label:
  a.append(i)
train_label = np.asarray(a)
```

```python
vocab_size = 10000 
dimesions = 16
```

```python
tf.keras.backend.clear_session()
model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(vocab_size , dimesions ),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(32 , return_sequences = True)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Bidirectional(tf.keras.layers.LSTM(64)),
    tf.keras.layers.Dropout(0.2),
    tf.keras.layers.Dense(512 , activation = 'relu'),
    tf.keras.layers.Dense(3 , activation = 'softmax')
])

model.summary()
```

```python
model.compile(loss  = tf.keras.losses.categorical_crossentropy, optimizer =  tf.keras.optimizers.Adam(), metrics = ['accuracy'])
```

```python
es_ = tf.keras.callbacks.EarlyStopping(monitor = 'accuracy' , patience=3)
```

```python
history = model.fit(padded_train_sequences , train_label , epochs = 4 , batch_size = 256 , validation_data = (padded_val_sequences , validation_label) ,callbacks = [es_])
```

```python
pred = model.predict(padded_test_sequences)
```
